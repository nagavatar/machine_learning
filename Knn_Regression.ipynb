# Using Machine Learning Algorithm i.e. KNN
We will be using knn from sklearn library which is most widely used library for data science.

```python
#importing StandardScaler function
#SS won't do scaling itself but python object will do scaling
from sklearn.preprocessing import StandardScaler
x_scaler=StandardScaler()
```

```python
#let's find the values of mean and std for 3 predictors using fit method
x_scaler.fit(df[['east','north','fl_area']])
#x_scaler.n_samples_seen_
#x_scaler.mean_
#x_scaler.var_
```

```python
#transforming x to z-values
X = x_scaler.transform(df[['east','north','fl_area']])

#alternative way to is to do fit and transform together 
#X=x_scaler.fit_transform(df[['east','north','fl_area']])
```

```python
print(X.shape)
#now we can see that the values are somewhat around 0 i.e. they are z-values
print(X[:6,:])
```

```python
#reason for using KNN is that since house price will be dependent on the locality and the nearby prices house so let's use that
#let's take 4 neighbors and keep weight as 'uniform' and take euclidean distance i.e. p=2 in Minnkowski distance
from sklearn.neighbors import KNeighborsRegressor as KNN
reg_object = KNN(n_neighbors=4,weights='uniform',p=2)
```

```python
price = df['price']/1000.0
#let's fit the model using standard X vs price to find the estimate of price.
reg_object.fit(X,price)
```

```python
#let's find the prediction
predicted_price = reg_object.predict(x_scaler.transform([[523800.0,179750.0,55.0]]))
print(predicted_price)
```

# Finding best Parameters
### Parameters in our case i.e. in KNN case
n i.e. nearest neighbor

p i.e. power of Minkowski distance

distance weighted mean or uniform mean

```python
#tuning the parameters so let's import evalutation metrics package and CV package
from sklearn.metrics import mean_absolute_error, make_scorer
#let's initialise mae object and gretr_is_better is FALSE tells that lower score is better
mae = make_scorer(mean_absolute_error, greater_is_better=False)
```

```python
#Just Specifying GridSearchCV for finding best parameters
from sklearn.model_selection import GridSearchCV
opt_nn = GridSearchCV(

    estimator = KNN(),
    scoring = mae,
    param_grid = {
        'n_neighbors':range(1,35),
        'weights':['uniform','distance'],
        'p':[1,2]})
```

```python
#let's do gridsearchCV for all 3 parameters
opt_nn.fit(X,price)
```

```python
#let's check the best estimator we got from above
print(opt_nn.best_estimator_.get_params())
```

```python
#predicting price using best parameters
predicted_hp = opt_nn.predict(
    x_scaler.transform([[523800.0,179750.0,55.0]]))
print(predicted_hp)
```

```python
#this is mean absolute error i.e. overall mean error of 26.4 thousand pounds
print(opt_nn.best_score_)
```

```python
#writing a function for much more beautiful o/p of best tuning parameters
def print_summary(opt_reg_object):
    params = opt_reg_object.best_estimator_.get_params()
    score = - opt_reg_object.best_score_
    print("Nearest neighbours: %8d" % params['n_neighbors'])
    print("Minkowski p       : %8d" % params['p'])
    print("Weighting         : %8s" % params['weights'])
    print("MAE Score         : %8.2f" % score)
    return
```

```python
print_summary(opt_nn)
```

# Let's do above thing using Pipelines
## Pipelines help us in doing above things like scaling, fitting and CV all at once


### Why Pipeline?
In ML project before fitting the model we need to do various transformations like encodding categorical variable, feature scaling, normalising. However, in every ML problem we need to put these transformations atleast twice first during training then during testing.

Ofcourse, we can do this using function but then also first we'll hv to call the function and then run the model separately.
Pipelines can do all these tasks at once. So, pipelines are used. Their main advantages are-
* Make the workflow simpler and easy to understand
* Enforce implementation and order of steps in Project
* Make code easily reproducible

```python
#importing Pipeline
from sklearn.pipeline import Pipeline
```

# Let's do all things which we done above using Pipeline also

```python
#creating Pipeline object
pipe = Pipeline([('zscores',StandardScaler()),
    ('NNreg',KNN(n_neighbors=4,weights='uniform',p=2))])
```

```python
#fitting the pipeline
pipe.fit(df[['east','north','fl_area']],price)
```

```python
#predicting using pipeline
print(pipe.predict([[523800.0,179750.0,55.0]]))
```

```python
#creating pipeline of zscores and KNN to find best parameters
pipe = Pipeline([('zscores',StandardScaler()),('NNreg',KNN())])
#giving pipeline as input in estimator of GRid serach CV
opt_nn2 = GridSearchCV(
    estimator = pipe,
    scoring = mae,
    param_grid = {
        'NNreg__n_neighbors':range(1,35),
        'NNreg__weights':['uniform','distance'],
        'NNreg__p':[1,2]})
```

```python
#fitting using created pipeline
opt_nn2.fit(df[['east','north','fl_area']],price)
```

```python
#predicting using best tuned pipeline
print(opt_nn2.predict([[523800.0,179750.0,55.0]]))
```

```python
opt_nn2.best_estimator_.get_params()
```

## Beautification function of best parameters

```python
#writing a function to get neater o/p of best paramters
def print_summary2(opt_pipe_object):
    params = opt_pipe_object.best_estimator_.get_params()
    score = - opt_pipe_object.best_score_
    print("Nearest neighbours: %8d" % params['NNreg__n_neighbors'])
    print("Minkowski p       : %8d" % params['NNreg__p'])
    print("Weighting         : %8s" % params['NNreg__weights'])
    print("MAE Score         : %8.2f" % score)
    return
```

```python
print_summary2(opt_nn2)
```

# Function to do housing prediction for any floor area given

```python
#alternatively we can do it using a function
#the advantage here is the convenience of changing the floor_area at single input
def surf3d(pipe_model,fl_area):
    east_mesh, north_mesh = np.meshgrid(
        np.linspace(505000,555800,100),
        np.linspace(158400,199900,100))
    fl_mesh = np.zeros_like(east_mesh)
    fl_mesh[:,:] = fl_area
    grid_predictor_vars = np.array([east_mesh.ravel(),
        north_mesh.ravel(),fl_mesh.ravel()]).T
    hp_pred = pipe_model.predict(grid_predictor_vars)
    hp_mesh = hp_pred.reshape(east_mesh.shape)
    fig = pl.figure()
    ax = Axes3D(fig)
    ax.plot_surface(east_mesh, north_mesh, hp_mesh, 
        rstride=1, cstride=1, cmap='YlOrBr',lw=0.01)
    ax.set_xlabel('Easting')
    ax.set_ylabel('Northing')
    ax.set_zlabel('Price at {} Floor Area'.format(fl_area))
    pl.show()
    return
```

```python
#plotting for floor area equal to 75
pl.close()
surf3d(opt_nn2,75.0)
pl.show()
```

```python
#plotting for floor area equal to 125
pl.close()
surf3d(opt_nn2,125.0)
pl.show()
```
